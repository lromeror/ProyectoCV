Creacion de columnas de longitudes y latitudes 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("assets/data/planificacion.xlsx",sheet_name="Hoja1")
df
```

```python
hola = set()
for coordenada in df['Coordenadas']:
    if type(coordenada)== float:
        print(coordenada)

```

```python
lats=[]
lons=[]
for coordenada in df['Coordenadas']:
    lats.append(coordenada.split(",")[0])
    lons.append(coordenada.split(",")[1])
df["LATS"] = lats
df["LONGS"] = lons
#df.to_excel("assets/data/planificacion.xlsx",sheet_name="Hoja1",index=False) 
df
```
